```python
#Create game code
def game_code():
    
    #Declare variable
    guesses_left = 7
    display = '_______'
    correct_word = 'stamina'
    guess_list = []
    
    #Main code block
    condition_main = False
    while condition_main is False:
        
        # Get user input and validate it is string
        condition = False
        while condition is False:
            guess = input('Guess a word or a letter: ').lower()
            if len(guess) == 0:
                print('You need to guess a letter or word')
            elif not guess.isalpha():
                print('Enter only a letter or a word')
            else:
                condition = True
        
        #Run checks to identify input is a letter or a word
        
        #scenario with letter
        if len(guess) == 1:
            
            #check if the letter has been already guessed
            if guess in guess_list:
                print('You already guessed this letter. Try again')
            else:
                #check if letter is part of targert word
                if guess in correct_word:
                    letter_indices = [index for index, char in enumerate(correct_word) if char == guess]
                    display = ''.join(guess if index in letter_indices else char for index, char in enumerate(display))
                    print("Great job. You guessed the right letter.")
                    print(f'{display} is your target word')
                    guesses_left-=1
                    guess_list.append(guess)
                    if display == correct_word:
                        print('You won!')
                        condition_main = True
                    else:
                        continue
                else:
                    #if letter is not part of target word
                    guesses_left-=1
                    if guesses_left == 0:
                        print('Your guess is wrong. You have 0 guesses left')
                        print('You lost')
                        condition_main = True
                    else:
                        print('Your guess is wrong. Try again')
                        print(f'You have {guesses_left} guesses left')
                        guess_list.append(guess)   
        
        #scenario with word
        elif len(guess) > 1:
            #check if the word has already been guessed
            if guess in guess_list:
                print('You already guessed this word. Try again')
            else:
                #check if the word is target word
                if guess == correct_word:
                    display = correct_word
                    print(f'{display} is your target word')
                    print('You won!')
                    condition_main = True                    
                else:
                    #if word is not target word
                    guesses_left-=1
                    if guesses_left ==0:
                        print('Your guess is wrong. You have 0 guesses left')
                        print('You lost')
                        condition_main = True
                    else:
                        print("Your guess is wrong. Try again")
                        print(f'You have {guesses_left} guesses left')
                        guess_list.append(guess)
```
